```python
# 1. importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# 2. Salary Data Real 2023 exploration
url = "https://raw.githubusercontent.com/ajcr/ajcr.github.io/master/ds_salaries.csv"
df = pd.read_csv(url)
```

```python

# 3. Filter for Spain + EU Remote Jobs      
df_spain = df[
    (df['company_location'] == 'ES') | 
    (df['employee_residence'] == 'ES') |
    (df['remote_ratio'] == 100) & (df['company_location'].isin(['GB','DE','NL','FR']))
]
```

```python

# 4. Calculate average salary by job title in Spain in EUR
salary_es = df_spain.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False) * 0.92
salary_es = salary_es.head(10) # Top 10 Salaries
```

```python

# 5. Plotting   
plt.figure(figsize=(12, 6))
sns.barplot(x=salary_es.values, y=salary_es.index, palette='viridis')
plt.title('Top 10 Data Jobs Salaries in Spain & EU Remote 2023', fontsize=16)
plt.xlabel('Average Salary in EUR')
plt.ylabel('Job Title')
plt.tight_layout()
plt.savefig('spain_salaries.png') # Save the plot as an image
plt.show()

```
